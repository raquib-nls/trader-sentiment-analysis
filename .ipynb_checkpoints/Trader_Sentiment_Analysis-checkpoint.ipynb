## Create the key metrics you will analyze, for example:
1) daily PnL per trader (or per account)
2) win rate, average trade size
3) leverage distribution
4) number of trades per day
5) long/short ratio


```python
daily_pnl = (
    merged_df
    .groupby(['Account', 'Date'])['Closed PnL']
    .sum()
    .reset_index()
)

daily_pnl.rename(columns={'Closed PnL': 'Daily_PnL'}, inplace=True)

daily_pnl.head()
```

```python
merged_df['Win'] = merged_df['Closed PnL'] > 0
```

```python
win_rate = (
    merged_df
    .groupby(['Account', 'Date'])['Win']
    .mean()
    .reset_index()
)

win_rate.rename(columns={'Win': 'Win_Rate'}, inplace=True)

win_rate.head()
```

```python
avg_trade_size = (
    merged_df
    .groupby(['Account', 'Date'])['Size USD']
    .mean()
    .reset_index()
)

avg_trade_size.rename(columns={'Size USD': 'Avg_Trade_Size'}, inplace=True)

avg_trade_size.head()
```

```python
trade_count = (
    merged_df
    .groupby(['Account', 'Date'])
    .size()
    .reset_index(name='Trade_Count')
)

trade_count.head()
```

```python
long_short = (
    merged_df
    .groupby(['Date', 'Side'])
    .size()
    .unstack(fill_value=0)
)

long_short['Long_Short_Ratio'] = long_short['BUY'] / (long_short['SELL'] + 1)

long_short = long_short.reset_index()

long_short.head()
```

```python
# combining
```

```python
metrics_df = daily_pnl.merge(win_rate, on=['Account','Date'])
metrics_df = metrics_df.merge(avg_trade_size, on=['Account','Date'])
metrics_df = metrics_df.merge(trade_count, on=['Account','Date'])

# Add sentiment
metrics_df = metrics_df.merge(
    merged_df[['Date','classification']].drop_duplicates(),
    on='Date',
    how='left'
)

metrics_df.head()
```

Engineered key behavioral and performance metrics including daily PnL, win rate, trade frequency, average trade size, and directional bias (long/short ratio).
These metrics enable comparison of trader behavior across different market sentiment regimes.

# Part B — Analysis

### 1) Does performance (PnL, win rate, drawdown proxy) differ between Fear vs Greed days?

```python
performance_by_sentiment = (
    metrics_df
    .groupby('classification')
    .agg({
        'Daily_PnL': 'mean',
        'Win_Rate': 'mean',
        'Trade_Count': 'mean'
    })
    .reset_index()
)

performance_by_sentiment
```

Interpretation

Highest profitability occurs during Fear (₹5328 avg PnL).

## Insight #1

Traders appear to perform better during Fear regimes compared to Greed regimes, suggesting contrarian or mean-reversion strategies may be more effective in pessimistic markets.

## Do traders change behavior based on sentiment?


```python
freq_by_sentiment = (
    metrics_df
    .groupby('classification')['Trade_Count']
    .mean()
    .reset_index()
)

freq_by_sentiment
```

Trading frequency increases significantly during Fear and Extreme Fear regimes, indicating that traders respond to uncertainty with heightened activity, likely attempting to exploit volatility.
In contrast, Greed periods exhibit lower trading intensity, suggesting more cautious participation and fewer opportunistic trades once bullish sentiment is already priced in.

```python
size_by_sentiment = (
    metrics_df
    .groupby('classification')['Avg_Trade_Size']
    .mean()
    .reset_index()
)

size_by_sentiment
```

Trading activity peaks during Extreme Fear, with traders executing significantly more trades and deploying larger capital per position compared to Greed regimes. This indicates that participants become more reactive and opportunity-seeking during volatile, pessimistic markets, attempting to exploit rapid price movements and perceived undervaluation. In contrast, Extreme Greed is associated with smaller trade sizes and reduced activity, suggesting more cautious participation despite optimistic sentiment.



```python
long_short_behavior = (
    merged_df
    .groupby(['classification', 'Side'])
    .size()
    .unstack(fill_value=0)
)

long_short_behavior['Long_Short_Ratio'] = (
    long_short_behavior['BUY'] / (long_short_behavior['SELL'] + 1)
)

long_short_behavior.reset_index()
```


During Extreme Fear, traders slightly favor long positions, possibly anticipating rebounds.
However, during Extreme Greed, the ratio drops below 1, indicating relatively more short trades — suggesting profit-taking or mean-reversion positioning.

Traders do not blindly follow sentiment; instead, they often position against extremes, reflecting strategic rather than emotional trading.

### Segment 1 — Frequent vs Infrequent Traders

```python
activity = metrics_df.groupby('Account')['Trade_Count'].mean()
threshold = activity.median()

metrics_df['Trader_Frequency'] = metrics_df['Account'].map(
    lambda x: 'Frequent' if activity[x] > threshold else 'Infrequent'
)
```

```python
freq_segment_perf = (
    metrics_df
    .groupby('Trader_Frequency')
    .agg({
        'Daily_PnL': 'mean',
        'Win_Rate': 'mean',
        'Trade_Count': 'mean',
        'Avg_Trade_Size': 'mean'
    })
    .reset_index()
)

freq_segment_perf
```

Frequent traders trade much more often and achieve higher profitability and win rates, indicating active strategies benefit from market participation.
Infrequent traders trade less and earn lower returns, suggesting limited engagement reduces opportunity capture.

#### Frequent traders exhibit higher trading activity and greater performance variability, particularly during Fear regimes, suggesting volatility-driven engagement.

## Segment 2 - Consistent Winners vs Inconsistent Traders

```python
consistency = metrics_df.groupby('Account')['Win_Rate'].mean()
threshold_win = consistency.median()

metrics_df['Trader_Consistency'] = metrics_df['Account'].map(
    lambda x: 'Consistent' if consistency[x] > threshold_win else 'Inconsistent'
)
```

```python
consistency_segment_perf = (
    metrics_df
    .groupby('Trader_Consistency')
    .agg({
        'Daily_PnL': 'mean',
        'Win_Rate': 'mean',
        'Trade_Count': 'mean',
        'Avg_Trade_Size': 'mean'
    })
    .reset_index()
)

consistency_segment_perf
```

Consistent traders achieve stronger win rates and higher average PnL with moderate trade sizes, reflecting disciplined execution.
Inconsistent traders take larger positions but generate weaker results, suggesting poor timing or risk control.

#### Consistent traders maintain superior win rates and more stable profitability across sentiment conditions, indicating disciplined execution compared to inconsistent participants

## Model Creation

```python
model_df = metrics_df.sort_values(['Account', 'Date'])

model_df['Next_Day_PnL'] = (
    model_df.groupby('Account')['Daily_PnL'].shift(-1)
)
```

```python
model_df['Target'] = (model_df['Next_Day_PnL'] > 0).astype(int)
```

```python
model_df = pd.get_dummies(model_df, columns=['classification'], drop_first=True)
```

```python
features = [
    'Trade_Count',
    'Avg_Trade_Size',
    'Win_Rate'
] + [col for col in model_df.columns if 'classification_' in col]

X = model_df[features]
y = model_df['Target']


X = X[:-1]
y = y[:-1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

preds = model.predict(X_test)

accuracy = accuracy_score(y_test, preds)
print("Model Accuracy:", accuracy)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, shuffle=False
)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

accuracy = model.score(X_test, y_test)
print("New Accuracy:", accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

rf_accuracy = rf.score(X_test, y_test)
print("Random Forest Accuracy:", rf_accuracy)
```

```python
importance = pd.Series(model.coef_[0], index=X.columns)
importance.sort_values().plot(kind='barh', figsize=(6,4))
```

# clustering Part

```python
cluster_df = metrics_df.groupby('Account').agg({
    'Trade_Count': 'mean',
    'Avg_Trade_Size': 'mean',
    'Win_Rate': 'mean',
    'Daily_PnL': 'mean'
}).reset_index()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_cluster = scaler.fit_transform(cluster_df.drop('Account', axis=1))
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
cluster_df['Cluster'] = kmeans.fit_predict(X_cluster)
```

```python
cluster_summary = (
    cluster_df
    .groupby('Cluster')
    [['Trade_Count', 'Avg_Trade_Size', 'Win_Rate', 'Daily_PnL']]
    .mean()
)

cluster_summary
```

### Clustering analysis revealed three distinct trader archetypes:
1) Active consistent traders who trade frequently with strong win rates and stable returns,
2) Low-engagement traders with limited participation and weaker performance, and
3) High-risk traders deploying significantly larger capital, achieving outsized profitability despite only moderate accuracy.
This highlights that profitability is driven by different behavioral styles rather than a single universal strategy.
